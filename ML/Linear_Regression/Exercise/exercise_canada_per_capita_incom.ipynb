```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('E:\\DataSets\\canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Income (USD$)')
plt.title("Canada's Per Capita Income")
plt.scatter(df.year, df.income)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.income)
```

```python
reg.predict([[2020]])
```

```python
year_df = pd.read_csv('E:\\DataSets\\canada_per_capita_predictions.csv')
year_df
```

```python
pred = reg.predict(year_df)
pred
```

```python
year_df['predicted_income'] = pred
year_df
```

```python
year_df.to_csv('E:\\DataSets\\canada_per_capita_predictions.csv')
```
