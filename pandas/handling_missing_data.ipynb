```python
import pandas as pd
df = pd.read_csv('E:\\DataSets\\weather_data1.csv', parse_dates=['day'])
df.set_index('day',inplace=True) # make day column as index
df
```

```python
new_df = df.fillna(0) # fill all NaN values with zero
new_df
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'  # filling different values of NaN for different columns using dictionary
})
new_df
```

```python
new_df = df.fillna(method='ffill') # this will carry forward the previous value.'bfill' means put previous value
new_df
```

```python
new_df= df.fillna(method='bfill',axis='columns') # fills the previous value from the adjacent column
new_df
```

```python
new_df = df.fillna(method='ffill', limit = 1) # fills the forward value only once
new_df
```

```python
new_df = df.interpolate() # it does the linear interpolation (linear is bydefault)
new_df
```

```python
new_df = df.interpolate(method='time')
new_df
```
