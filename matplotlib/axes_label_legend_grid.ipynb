# Axes labels, Legends, Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days =  [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days,max_t, label='Max') # to activate legends you must first label each plot.
plt.plot(days,min_t, label ='Min')
plt.plot(days,avg_t, label = 'Avg')

plt.legend(loc='best', shadow=True, fontsize='small') # this will show the labels of each plot (legeng)
#loc can be 'upper right', 'best'. bydefault shadow is flase.

plt.grid() # it will show plot on grid.
```
