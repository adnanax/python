```python
plt.xticks(xpos,company) 
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue, width=0.4, label = 'Revenue') #control the width of the bars
plt.bar(xpos+0.2,profit, width=0.4, label = 'Profit') 
plt.legend()
```

```python
plt.yticks(xpos,company) #company names of be on the yaxis
plt.title('US Tech Stocks')
plt.barh(xpos-0.2,revenue, label = 'Revenue') #barh used for horizontal bar
plt.barh(xpos+0.2,profit, label = 'Profit') 
plt.legend()
```
