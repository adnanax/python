# Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,200,410,250]
exp_labels = ['House Rent','Food','Phone/Internet Bill', 'Car','Other Utilities']
```

```python
plt.axis('equal') # to make the pie chart a perfect circle, used when it is otherwise
plt.pie(exp_vals, labels = exp_labels)
```

```python
plt.axis('equal') 
plt.pie(exp_vals, labels = exp_labels)
plt.show()  # to remove all the extra labels
```

```python
plt.axis('equal') 
plt.pie(exp_vals, labels = exp_labels, radius=1.5) # radius is used to select the size of the circle(bdflt it is 1)
plt.show()
```

```python
plt.axis('equal') 
plt.pie(exp_vals, labels = exp_labels, radius=1.5, autopct='%0.1f%%')#the value between '%f%%', specifies the number decimal
plt.show()                                                           # places you want to show after percentage 0.2 mean 2 
                                                                    #decimal place after percentage.
```

```python
plt.axis('equal') 
plt.pie(exp_vals, labels = exp_labels, radius=1.5, autopct='%0.1f%%', shadow= True) #for shadows
plt.show()
```

```python
plt.axis('equal') 
plt.pie(exp_vals, labels = exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0.2,0,0,0]) #to explode the
plt.show()                                                                                  #specific piece from the chart
```

```python
plt.axis('equal') 
plt.pie(exp_vals, labels = exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0.2,0,0,0],startangle=45)
plt.show()   #mention the degree of the angle to start the pie chart e.g; 30,45,90,180 etc...
```
