# Format Strings in plot function

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y ,'g+')
```

```python
plt.plot(x,y ,'g+--plt.plot(x,y ,'g+')')
```

```python
plt.plot(x,y ,'r*--')
```

```python
plt.plot(x,y ,'rd')
```

```python
plt.plot(x,y, color='red', marker='D', linestyle='')
```

```python
plt.plot(x,y, color='0100008', marker='D', linestyle='')# also give RGB(000000) hexadecimal code for color
```

```python
plt.plot(x,y, color='red', marker='D', linestyle='', markersize=20)
```

```python
plt.plot(x,y, color='green', marker='+', alpha=0.6)# alpha controls the trasparency of the color(0-1) 0=transparent 1=translucent
```
